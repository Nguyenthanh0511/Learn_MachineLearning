## Bước 1: Tải tập dữ liệu Adult

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read data from file 
adult_data = pd.read_csv('data/adult_cleaned.csv')

# Some informations of adult dataset
print(adult_data.shape)
print(adult_data.info())
adult_data.head()

# Count the number of elements in each class
income_counts = adult_data['income'].value_counts()
plt.pie(income_counts, labels=income_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Income')

plt.show()
```
