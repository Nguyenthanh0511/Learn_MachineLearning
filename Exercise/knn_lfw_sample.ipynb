## Bước 1: Tải tập dữ liệu LFW

```python
# Import thư viện cần thiết
from sklearn.datasets import fetch_lfw_people  # Thư viện để tải tập dữ liệu LFW (Labeled Faces in the Wild)
import matplotlib.pyplot as plt  # Thư viện để vẽ đồ thị và biểu đồ
import numpy as np
import pandas as pd

# Tải dữ liệu LFW với điều kiện chỉ lấy những người có ít nhất 70 hình ảnh, resize hình ảnh xuống 40% kích thước gốc
lfw_dataset = fetch_lfw_people(min_faces_per_person=70, resize=0.4)

# Lấy dữ liệu hình ảnh từ tập dữ liệu
imgs = lfw_dataset.images

labels = lfw_dataset.target

# Lấy tên các lớp (tên của những người trong tập dữ liệu)
target_names = lfw_dataset.target_names

# Số lượng lớp 
n_classes = target_names.shape[0]

# In ra thông tin về tập dữ liệu
print('Số lượng mẫu trong tập dữ liệu: {}'.format(imgs.shape[0]))  # Số lượng hình ảnh trong tập dữ liệu
print('Kích thước của hình ảnh: {}'.format(imgs.shape[1:]))  # Kích thước của từng hình ảnh (height, width)
print('Số lượng lớp: {}'.format(n_classes))  # Số lượng lớp (số người khác nhau)

# Hiển thị một vài hình ảnh từ tập dữ liệu
fig, axes = plt.subplots(2, 5, figsize=(15, 8))  # Tạo một grid 2x5 để hiển thị 10 hình ảnh
for i, ax in enumerate(axes.flat):  # Duyệt qua từng ô trong grid
    ax.imshow(lfw_dataset.images[i], cmap='gray')  # Hiển thị hình ảnh ở dạng grayscale
    ax.set_xticks([])  # Loại bỏ tick trên trục x
    ax.set_yticks([])  # Loại bỏ tick trên trục y
    ax.set_title(lfw_dataset.target_names[lfw_dataset.target[i]])  # Đặt tiêu đề là tên của người trong hình
plt.show()  # Hiển thị các hình ảnh trên grid

# Số lượng hình ảnh của mỗi lớp (mỗi người)
unique, counts = np.unique(labels, return_counts=True)
class_counts = dict(zip(target_names[unique], counts))

# Chuyển đổi sang DataFrame để hiển thị
class_counts_df = pd.DataFrame(list(class_counts.items()), columns=['Name', 'Number of Images'])

# Hiển thị số lượng hình ảnh của mỗi lớp
print(class_counts_df)
```
